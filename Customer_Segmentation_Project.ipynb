# Customer Segmentation using K-Means
Dataset: 12,000+ customer records

```
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

df = pd.read_csv('customer_segmentation_12000.csv')
df.head()
```

```
df.shape
```

```
features = df[['Age','Annual_Income','Spending_Score']]
scaler = StandardScaler()
X = scaler.fit_transform(features)
```

```
wcss=[]
for i in range(1,11):
    km = KMeans(n_clusters=i, random_state=42, n_init=10)
    km.fit(X)
    wcss.append(km.inertia_)
plt.plot(range(1,11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Clusters')
plt.ylabel('WCSS')
plt.show()
```

```
kmeans = KMeans(n_clusters=5, random_state=42, n_init=10)
df['Cluster'] = kmeans.fit_predict(X)
df.head()
```

```
plt.scatter(df['Annual_Income'], df['Spending_Score'], c=df['Cluster'])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Customer Clusters')
plt.show()
```

```
cluster_profile = df.groupby('Cluster')[['Age','Annual_Income','Spending_Score']].mean()
cluster_profile
```

## Conclusion
